```python
import random
import time

# Define the card deck using a list of dictionaries
card_deck = [
    {"name": "Diablo", "health": 100, "attack": 90, "defense": 60},
    {"name": "Medusa", "health": 100, "attack": 70, "defense": 70},
    {"name": "Jester", "health": 120, "attack": 60, "defense": 90},
    {"name": "Troll", "health": 150, "attack": 40, "defense": 94},
    {"name": "Specter", "health": 100, "attack": 70, "defense": 70},
    {"name": "Mist", "health": 100, "attack": 75, "defense": 65},
    {"name": "Savage", "health": 100, "attack": 90, "defense": 50},
    {"name": "Marauder", "health": 100, "attack": 85, "defense": 50},
    {"name": "Wimp", "health": 110, "attack": 40, "defense": 85},
    {"name": "Sorcerer", "health": 100, "attack": 70, "defense": 55}
]

# Number of cards for each player
deck_size = 5

# Draw random cards for the player and opponent
player_deck = random.sample(card_deck, deck_size)
opponent_deck = random.sample(card_deck, deck_size)

def show_card(card):
    print(f"Name: {card['name']}, Health: {card['health']}, Attack: {card['attack']}, Defense: {card['defense']}")

def attack(attacker, defender):
    # If attack value is greater than defender's health, defender is defeated
    if attacker['attack'] > defender['health']:
        print(f"{attacker['name']} defeats {defender['name']} with a powerful attack!")
        return True
    
    # Calculate damage based on defender's defense
    damage = max(0, attacker['attack'] - defender['defense'])
    if defender['defense'] > 0:
        defender['defense'] = max(0, defender['defense'] - attacker['attack'])
        print(f"{defender['name']}'s defense reduced to {defender['defense']}!")
    
    if defender['defense'] == 0 and damage > 0:
        defender['health'] -= damage
        print(f"{defender['name']} takes {damage} damage, health is now {defender['health']}!")
    
    if defender['health'] <= 0:
        print(f"{defender['name']} is defeated!")
        return True
    return False

def player_turn():
    print("\nYour turn!")
    print("Your deck:")
    for idx, card in enumerate(player_deck):
        print(f"{idx + 1}.", end=" ")
        show_card(card)

    opponent_card = random.choice(opponent_deck)
    print("\nOpponent plays:")
    show_card(opponent_card)

    choice = int(input("Choose a card to attack with (1-5): ")) - 1
    player_card = player_deck[choice]

    print(f"\nYou attack with {player_card['name']}!")
    if attack(player_card, opponent_card):
        opponent_deck.remove(opponent_card)

def opponent_turn():
    print("\nOpponent's turn!")
    time.sleep(2)
    
    opponent_card = random.choice(opponent_deck)
    player_card = random.choice(player_deck)
    
    print(f"Opponent attacks with {opponent_card['name']}")
    show_card(player_card)
    
    if attack(opponent_card, player_card):
        player_deck.remove(player_card)

# Main game loop
while player_deck and opponent_deck:
    player_turn()
    if not opponent_deck:
        print("You defeated all opponent cards. You win!")
        break
    
    time.sleep(2)
    
    opponent_turn()
    if not player_deck:
        print("The opponent defeated all your cards. You lose!")
        break
    
    print("\nA new round begins in 3 seconds...\n")
    time.sleep(3)

print("Game over.")
```
